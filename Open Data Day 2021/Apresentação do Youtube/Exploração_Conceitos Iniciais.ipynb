## __Régua de Correlação__

|Range de Valores | Descrição|
|---------------:|:--------:|
|__-1.0 ~ -0.5__ |Correlação forte inversa. Quando um sobe o outro desce. |
|__-0.5 ~ 0.0__ |Correlação fraca inversa. Quando um sobe o outro desce mas não tem uma alta interferencia.|
|__0.0 ~ 0.5__ |Correlação fraca direta. Quando um sobe o outro sobre mas não com uma alta.|
|__0.5 ~ 1.0__ |Correlação forte direta. Quando um sobre o outro sobre também!|

__Logo .... quanto mais próximo do 1, temos uma correlação forte positiva, quanto mais próximo do -1, temos uma correlação forte negativa.__

* __HeatMap__ --> https://seaborn.pydata.org/examples/many_pairwise_correlations.html

```python
corr = df_vacina.corr().round(4)
corr = corr[corr > 0]

mask = np.triu(np.ones_like(corr, dtype=bool))

fig, ax = plt.subplots(figsize=(11, 9))

ax = sns.heatmap(corr, mask=mask, cmap='Blues',
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

ax.set_title("Gráfico de Correlação das Variavéis", fontdict={'fontsize': 16, 'color': 'Blue'})
```

## __Corrigindo Strings Similares__

<br>__Em teoria da informação , linguística e ciência da computação , a distância de Levenshtein é uma métrica de string para medir a diferença entre duas sequências. Informalmente, a distância Levenshtein entre duas palavras é o número mínimo de edições de um único caractere (inserções, exclusões ou substituições) necessárias para transformar uma palavra na outra. É nomeado após o matemático soviético Vladimir Levenshtein , que considerou esta distância em 1965.__

> O índice de similaridade vai de 0 a 100

* [Função de verificar caracteres](https://stackoverflow.com/questions/60287306/how-to-check-if-string-contains-chars-which-are-not-in-a-list)
* [Wikipedia Levenshtein distance](https://en.wikipedia.org/wiki/Levenshtein_distance)

```python
lista_cidades = list(df_vacina['MUNICIPIO'].unique())
```

```python
lista_cidades
```

```python
import string

filtro = frozenset(
    string.ascii_letters 
    + string.digits
    + '-._' + ' ' + "'"+'ç' +'á'+'à'+ 'ã'+ 'â'+'é'+ 'è'+ 'ê'+ 'í'+ 'ì'+ 'î'+'ó'+ 'ó'+ 'õ'+ 'ô'+'ú'+'ù'+ 'û'
)

def verifica_caracter_especial(string, filtro=filtro):
    return not set(string) <= filtro
```

```python
#TalismÆ
verifica_caracter_especial("TalismÆ")
```

```python
municipio_errado = []
municipio_correto = []

for i in lista_cidades:
    if verifica_caracter_especial(i.lower()):
        municipio_errado.append(i)
    else:
        municipio_correto.append(i)

print("Temos {0} municipios preenchidos de forma errada e {1} preenchidos corretamente.".format(
    len(municipio_errado),len(municipio_correto)))
```

## __Validando os resultados encontrados__

__Exemplo:__ 

O municipio Palmeiras do Tocantins teve um match com  Para¡so do Tocantins de 86 pontos.
<br> Mas os dois municipios existem, apenas as letras que tem uma correlação alta devido ao __"do Tocantins"__, por ambas começarem com __"P"__

```python
for correto in municipio_correto:
    for errado in municipio_errado:
        if 85 <= fuzz.ratio(correto, errado) <= 90 :
            print("{0} --> {1} = {2}".format(correto, errado, fuzz.ratio(correto, errado)))
```

### __Atualizando o DataFrame__

```python
for correto in municipio_correto:
    
    for index, errado in df_vacina[df_vacina['MUNICIPIO'].isin(municipio_errado)].iterrows():
        
        if fuzz.ratio(correto, errado['MUNICIPIO']) >= 90:
            
            df_vacina.loc[df_vacina['MUNICIPIO'] == errado['MUNICIPIO'], ['MUNICIPIO']] = correto
```
