# Árvores de Decisão ou Desion Tree


Como o algoritimo funciona:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('..\Base de Dados\credit.pkl', mode = 'rb') as f:
    X_credit_train, y_credit_train, X_credit_test, y_credit_test = pickle.load(f)
```

```python
X_credit_train.shape, y_credit_train.shape
```

```python
lst_ = []
for a in y_credit_train:
    lst_.append(a[0])

y_credit_train = np.array(lst_)
```

```python
X_credit_train.shape, y_credit_train.shape
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
lst_ = []
for a in y_credit_test:
    lst_.append(a[0])

y_credit_test = np.array(lst_)
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
arvore_credit = DecisionTreeClassifier(criterion = 'entropy')
```

```python
arvore_credit.fit(X_credit_train, y_credit_train)
```

```python
# 
arvore_credit.feature_importances_
```

```python
from sklearn import tree
```

```python
tree.plot_tree(arvore_credit)
```

```python
previsores = ['income', 'age', 'loan']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,20))

tree.plot_tree(
    arvore_credit, 
    feature_names = previsores,
    class_names = ['0', '1'], #arvore_credit.classes_
    filled = True
);

fig.savefig('arvore_credito.png')
```

```python
X_credit_test[0], y_credit_test[0]
```

```python
arvore_credit.predict([[26680.14583644,    47.76173998,  1671.18492392]])
```

```python
previsoes = arvore_credit.predict(X_credit_test)
previsoes
```

```python
y_credit_test
```

```python
accuracy_score(y_credit_test, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(arvore_credit)
cm.fit(X_credit_train, y_credit_test)
cm.score(X_credit_test, y_credit_test)
```

```python
print( classification_report(y_credit_test, previsoes) )
```

```python
# Classe 0
# O algoritimo consegue identificar 99% dos registros da classe 0.
# Quando ele identifica um registro dessa classe, a precisão é de 99%.

# Classe 1
# O algoritimo consegue identificar 96% dos registros da classe 1.
# Quando ele identifica um registro dessa classe, a precisão é de 94%
```

# Florestas Randômicas ou Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_credit_train.shape, y_credit_train.shape
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
random_forest_credit = RandomForestClassifier(
    n_estimators = 40, #numero de arvores que serao criadas
    criterion = 'entropy', #modelo de calculo
    random_state = 0 #numero aleatorio para ter o mesmo resultado
)
```

```python
# Treina o modelo com a base de treino
random_forest_credit.fit(X_credit_train, y_credit_train)
```

```python
# gera as porevisoes com a base de teste

previsoes = random_forest_credit.predict(X_credit_test)
previsoes
```

```python
y_credit_test
```

```python
accuracy_score(y_credit_test, previsoes)
```

```python
cm = ConfusionMatrix(random_forest_credit)
cm.fit(X_credit_train, y_credit_test)
cm.score(X_credit_test, y_credit_test)
```

```python
print( classification_report(y_credit_test, previsoes) )
```
